```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)
```

```python
filepath = 'C:/Users/night/OneDrive/Documents/Out of the Park Developments/OOTP Baseball 22/saved_games/SBC.lg/import_export/csv/'
managers = pd.read_csv(filepath + 'human_managers.csv')
record_history = pd.read_csv(filepath + 'human_manager_history_record.csv')
manger_history = pd.read_csv(filepath + 'human_manager_history.csv')
```

```python
managers = managers[['human_manager_id','first_name', 'last_name', 'retired', 'experience']]
history = record_history.merge(manger_history, how = 'inner', on = ('human_manager_id', 'team_id', 'year'))
history = history[['human_manager_id','year', 'g', 'w', 'l', 'pos','made_playoffs', 'won_playoffs']]
final_managers = history.merge(managers, how = 'left', on = ('human_manager_id'))
final_managers['name'] = final_managers['first_name'] + ' ' + final_managers['last_name']
```

```python
summary_cols = ['name', 'experience','g','w','l','pos','made_playoffs','won_playoffs']
sum_stats = final_managers[summary_cols].groupby(['name', 'experience']).sum().reset_index(drop=False).sort_values('g', ascending = False)
sum_stats = sum_stats.groupby(['name']).sum().reset_index(drop=False)
sum_stats['Win %'] = round((sum_stats['w']/sum_stats['g'])*100,1)
sum_stats['Avg Finish'] = round(sum_stats['pos'] / sum_stats['experience'],1)
sum_stats['Record'] = sum_stats['w'].map(str) + '-' + sum_stats['l'].map(str)
sum_stats['Games'] = sum_stats['g']
sum_stats['GM'] = sum_stats['name']
sum_stats['Seasons'] = sum_stats['experience']
sum_stats['Playoffs'] = sum_stats['made_playoffs']
sum_stats['Titles'] = sum_stats['won_playoffs']
sum_stats = sum_stats[sum_stats['experience'] >= 8]
sum_stats = sum_stats[sum_stats['name'] != 'Place Holder']


final_cols = ['GM', 'Seasons','Games', 'Record','Win %', 'Avg Finish', 'Playoffs', 'Titles']

sum_stats[final_cols].sort_values('Win %', ascending = False).set_index('GM')
```
